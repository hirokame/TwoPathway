```python
import pandas as pd
import h5py
import glob
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.cm as cm
from scipy import stats

from collections import Counter
from scipy.stats import ttest_rel
```

```python
posefiles_PNOC = glob.glob("D:/PNOC_opto/*filtered.csv")
posefiles_PNOC.sort()
for i in range(len(posefiles_PNOC)):
    print(posefiles_PNOC[i])
```

```python
posefiles_NTS = glob.glob("D:/NTS_opto/*filtered.csv")
posefiles_NTS.sort()
for i in range(len(posefiles_NTS)):
    print(posefiles_NTS[i])
```

```python
stimfiles_PNOC = glob.glob("D:/PNOC_opto/Stimulation/*csv")
stimfiles_PNOC.sort()
for i in range(len(stimfiles_PNOC)):
    print(stimfiles_PNOC[i])
```

```python
stimfiles_NTS = glob.glob("D:/NTS_opto/Stimulation/*csv")
stimfiles_NTS.sort()
for i in range(len(stimfiles_NTS)):
    print(stimfiles_NTS[i])
```

```python
PNOC_stims = []
for i in range(8):
    stim = pd.read_csv(stimfiles_PNOC[i])
    padding = np.ones(stim["Item1.Value"][0])
    stim = np.hstack((padding, np.array(stim["Item2"])))
    keep_idx = np.where(np.diff(np.append(stim,1))==0)
    stim = stim[keep_idx]
    
    pose = pd.read_csv(posefiles_PNOC[i])
    pose = pose[2:]
    
    print(len(stim))
    
    for j in range(len(stim)-1):
        if (stim[j+1]==0 and stim[j]==1):## True to False; stimulation start
            for k in range(240):
                stim[j-k]=2

    for j in range(len(stim)-1):
        if (stim[j+1]==1 and stim[j]==0):## False to True; stimulation end
            for k in range(240):
                stim[j+1+k]=3
    PNOC_stims.append(stim)
```

```python
NTS_stims = []
for i in range(11):
    stim = pd.read_csv(stimfiles_NTS[i])
    padding = np.ones(stim["Item1.Value"][0])
    stim = np.hstack((padding, np.array(stim["Item2"])))
    keep_idx = np.where(np.diff(np.append(stim,1))==0)
    stim = stim[keep_idx]
    
    pose = pd.read_csv(posefiles_NTS[i])
    pose = pose[2:]
    
    print(len(stim))
    
    for j in range(len(stim)-1):
        if (stim[j+1]==0 and stim[j]==1):## True to False; stimulation start
            for k in range(240):
                stim[j-k]=2
    for j in range(len(stim)-1):
        if (stim[j+1]==1 and stim[j]==0):## False to True; stimulation end
            for k in range(240):
                stim[j+1+k]=3
    NTS_stims.append(stim)
```

```python
PNOC_speed_Pre = []
PNOC_speed = []
PNOC_speed_Stim = []
PNOC_speed_Post = []
for i in range(len(posefiles_PNOC)):
    df = pd.read_csv(posefiles_PNOC[i])
    x = df['DLC_resnet50_small_open_field_Nov23Nov21shuffle1_910000.6'][2:].to_numpy(dtype=float)
    y = df['DLC_resnet50_small_open_field_Nov23Nov21shuffle1_910000.7'][2:].to_numpy(dtype=float)
    dx = np.diff(x)
    dy = np.diff(y)
    speed = np.sqrt(dx**2 + dy**2)*30
    speed = np.append(speed, 0)

    speed_Pre = speed[PNOC_stims[i]==2]
    PNOC_speed_Pre.append(np.average(speed_Pre))
    PNOC_speed.append(np.average(speed))
    speed_Stim = speed[PNOC_stims[i]==0]
    PNOC_speed_Stim.append(np.average(speed_Stim))
    speed_Post = speed[PNOC_stims[i]==3]
    PNOC_speed_Post.append(np.average(speed_Post))

NTS_speed_Pre = []
NTS_speed = []
NTS_speed_Stim = []
NTS_speed_Post = []
for i in range(len(posefiles_NTS)):
    df = pd.read_csv(posefiles_NTS[i])
    x = df['DLC_resnet50_small_open_field_Nov23Nov21shuffle1_910000.6'][2:].to_numpy(dtype=float)
    y = df['DLC_resnet50_small_open_field_Nov23Nov21shuffle1_910000.7'][2:].to_numpy(dtype=float)
    dx = np.diff(x)
    dy = np.diff(y)
    speed = np.sqrt(dx**2 + dy**2)*30
    speed = np.append(speed, 0)
    
    speed_Pre = speed[NTS_stims[i]==2]
    NTS_speed_Pre.append(np.average(speed_Pre))
    NTS_speed.append(np.average(speed))
    speed_Stim = speed[NTS_stims[i]==0]
    NTS_speed_Stim.append(np.average(speed_Stim))
    speed_Post = speed[NTS_stims[i]==3]
    NTS_speed_Post.append(np.average(speed_Post))
```

```python
plt.figure(figsize=(2,3))
plt.title("Speed Comp(Pre)")
sns.boxplot([PNOC_speed_Pre, NTS_speed_Pre])
sns.swarmplot([PNOC_speed_Pre, NTS_speed_Pre], color='k')
plt.xticks([0, 1], ['PNOC', 'NTS'])
plt.ylabel("Speed (pxls/sec)")
plt.tight_layout()
plt.ylim(0,250)
plt.savefig("SpeedComp_Pre.svg")
plt.savefig("SpeedComp_Pre.png")
t_stat_all, p_value_all = stats.ttest_ind(PNOC_speed_Pre, NTS_speed_Pre)
p_value_all
```

```python
plt.figure(figsize=(2,3))
plt.title("Speed Comp(Entire)")
sns.boxplot([PNOC_speed, NTS_speed])
sns.swarmplot([PNOC_speed, NTS_speed], color='k')
plt.xticks([0, 1], ['PNOC', 'NTS'])
plt.ylabel("Speed (pxls/sec)")
plt.tight_layout()
plt.ylim(0,250)
plt.savefig("SpeedComp_entire.svg")
plt.savefig("SpeedComp_entire.png")
t_stat_all, p_value_all = stats.ttest_ind(PNOC_speed, NTS_speed)
p_value_all
```

```python
plt.figure(figsize=(2,3))
plt.title("Speed Comp(Stim)")
sns.boxplot([PNOC_speed_Stim, NTS_speed_Stim])
sns.swarmplot([PNOC_speed_Stim, NTS_speed_Stim], color='k')
plt.xticks([0, 1], ['PNOC', 'NTS'])
plt.ylabel("Speed (pxls/sec)")
plt.tight_layout()
plt.ylim(0,250)
plt.savefig("SpeedComp_stim.svg")
plt.savefig("SpeedComp_stim.png")
t_stat_all, p_value_all = stats.ttest_ind(PNOC_speed_Stim, NTS_speed_Stim)
p_value_all
```

```python
plt.figure(figsize=(5,3))
plt.title("Speed Comparison")
sns.boxplot([PNOC_speed_Pre, NTS_speed_Pre, PNOC_speed_Stim, NTS_speed_Stim, PNOC_speed, NTS_speed], palette=['blue','orange','blue','orange','blue','orange'], linewidth=1)
sns.swarmplot([PNOC_speed_Pre, NTS_speed_Pre, PNOC_speed_Stim, NTS_speed_Stim, PNOC_speed, NTS_speed], color='k')
plt.xticks([0, 1,2,3,4,5], ['PNOC', 'NTS','PNOC', 'NTS','PNOC', 'NTS'])
plt.ylabel("Speed (pxls/sec)")
plt.tight_layout()
plt.savefig("SpeedComp_all.svg")
plt.savefig("SpeedComp_all.png")
t_stat_all, p_value_all = stats.ttest_ind(PNOC_speed_Pre, NTS_speed_Pre)
p_value_all
```
